```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns = ['CustomerID', 'Gender', 'Age', 'Income', 'SpendingScore']
df.head()
```

```python
df['Income_Group'] = pd.cut(
    df['Income'],
    bins=[0, 40, 70, 150],
    labels=['Low Income', 'Medium Income', 'High Income']
)
df['Income_Group']
```

```python
df['Spending_Group'] = pd.cut(
    df['SpendingScore'],
    bins=[0, 40, 70, 100],
    labels=['Low Spender', 'Medium Spender', 'High Spender']
)
df['Spending_Group']
```

```python
segment_summary = df.groupby(
    ['Income_Group', 'Spending_Group'],
    observed=False
).agg(
    Customers=('CustomerID', 'count'),
    Avg_Age=('Age', 'mean'),
    Avg_Spending=('SpendingScore', 'mean')
).reset_index()
segment_summary
```

```python
df.to_csv("segmented_customers.csv", index=False)
segment_summary.to_csv("segment_summary.csv", index=False)
```
